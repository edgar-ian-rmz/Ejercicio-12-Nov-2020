# Slicing continuación

[inicio: stop: incremento]

notas:

1. el stop no es incluido.
2. el incremento se puede modificar
3. soprta indexado alternativo
4.  

```
frase = "como hacer a la gente mas positiva?, muy facil, has que pierdan un electron!"
print(frase)
#print ace de la palabra hacer
print(frase [6 : 9 : 1])

#incremento valores por defecto
print(frase [ : 5 : 2 ])

#incremento
print(frase[::2])
print(frase[-1::-1])
```

```
#metodos para procesar stirng

numeros = " 2 , 1 ,3,4,    5   ,   45 ,  7     "
print(numeros)
#como estraigo los numeros de la cadena
#paso 1: eliminar los espacios al principio y al final con strip()
numeros = numeros.strip()
print(numeros)

#paso 2:metodo split
numeros = numeros.split(",")
print(numeros)

#paso 3: con un for e indices limpiar cadenas
for indice in range( len(numeros)):
  numeros[indice] = numeros[indice].strip()
print(numeros)

#paso 4: convertirlos a tipo antero
for indice in range( len(numeros)):
  numeros[indice] = int(numeros[indice])
print(numeros)

#paso 5: usar los numeros
```
